```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
date_list = pd.date_range(start='2025-01-01', end='2025-07-15', freq='D')
for date in date_list:
    print(f"Processing data for {date.strftime('%Y%m%d')}")
```

```python
# loop through bulk forecast
bulktrans_df = pd.DataFrame()
for date in date_list:
    file_path = f'CircuitLevelData/Bulk_Forecasts 1-1-25 through 7-15-25/circuit_level_weather_fpi2_BulkTrans_HFA_{date.strftime("%Y%m%d")}.csv'
    try:
        df = pd.read_csv(file_path)
        df['Date'] = date.strftime('%Y-%m-%d')
        bulktrans_df = pd.concat([bulktrans_df, df], ignore_index=True)
    except FileNotFoundError:
        print(f"File not found: {file_path}")
```

```python
# loop through dist forecast
dist_df = pd.DataFrame()
for date in date_list:
    file_path = f'CircuitLevelData/Dist_Forecasts 1-1-25 through 7-15-25/circuit_level_weather_fpi2_Distribution_HFA_{date.strftime("%Y%m%d")}.csv'
    try:
        df = pd.read_csv(file_path)
        df['Date'] = date.strftime('%Y-%m-%d')
        dist_df = pd.concat([dist_df, df], ignore_index=True)
    except FileNotFoundError:
        print(f"File not found: {file_path}")
dist_df = dist_df[pd.to_datetime(dist_df['Datetime']).dt.year == 2025]
dist_df
```

```python
# loop through sub forecast
sub_df = pd.DataFrame()
for date in date_list:
    file_path = f'CircuitLevelData/Sub_Forecasts 1-1-25 through 7-15-25/circuit_level_weather_fpi2_SubTrans_HFA_{date.strftime("%Y%m%d")}.csv'
    try:
        df = pd.read_csv(file_path)
        df['Date'] = date.strftime('%Y-%m-%d')
        sub_df = pd.concat([sub_df, df], ignore_index=True)
    except FileNotFoundError:
        print(f"File not found: {file_path}")
```

```python
cols_to_keep_trans = ['Datetime','Master Circuit','FuelsIndex_max_FPI2','LFPwtdd_max_FPI2']
cols_to_keep_dist = ['Datetime','Circuit','FuelsIndex_max_FPI2','LFPwtdd_max_FPI2']
forecast_circuit_2025 = pd.concat([bulktrans_df[cols_to_keep_trans].rename(columns={'Master Circuit':'Circuit Name'}), 
                                   dist_df[cols_to_keep_dist].rename(columns={'Circuit': 'Circuit Name'}), 
                                   sub_df[cols_to_keep_trans].rename(columns={'Master Circuit':'Circuit Name'})], ignore_index=True)
# Convert 'Datetime' to datetime type
forecast_circuit_2025['Datetime'] = pd.to_datetime(forecast_circuit_2025['Datetime'])
# make sure only 2025 data is kept
forecast_circuit_2025 = forecast_circuit_2025[forecast_circuit_2025['Datetime'].dt.year == 2025]
# rename columns
forecast_circuit_2025 = forecast_circuit_2025.rename(columns={
    'FuelsIndex_max_FPI2': 'max_max_fpi_2.0_tdd_fuels_index',
    'LFPwtdd_max_FPI2': 'max_max_lfpw_using_dew_point_depression'
},)
# create a string Date column
forecast_circuit_2025['Date'] = forecast_circuit_2025['Datetime'].dt.strftime('%Y-%m-%d')
# group by Date, Circuit Name, and Circuit_No, set max_max_fpi_2.0_tdd_fuels_index and max_max_lfpw_using_dew_point_depression to the max value
forecast_circuit_2025 = forecast_circuit_2025.groupby(['Date', 'Circuit Name', ]).agg({
    'max_max_fpi_2.0_tdd_fuels_index': 'max',
    'max_max_lfpw_using_dew_point_depression': 'max'
}).reset_index()
```

```python
# create a list of Date ranges from 07-07 to today date
today = pd.to_datetime("today").strftime("%Y-%m-%d")
date_ranges = pd.date_range(start='2025-07-16', end=today)
# create a list of valid dates
valid_dates = ['07-16-2025']
# create an empty dataframe for forecast data
forecast_file_group = pd.DataFrame()
# loop through date ranges
for date in date_ranges:
    # try to read the forecast file
    try:
        print(date.strftime('%m-%d-%Y'))
        forecast_file = pd.read_csv(f"//sce.eix.com/workgroup/SPE_DSA/Charles/ADS_forecast/Dist Circuit FPI2 Forecast_{date.strftime('%m-%d-%Y')}.csv",)
        # make sure date is in string format
        forecast_file['Date'] = pd.to_datetime(forecast_file['Datetime']).dt.strftime('%Y-%m-%d')
        # keep only current date
        forecast_file = forecast_file[forecast_file['Date'] == date.strftime('%Y-%m-%d')]
        # append to forecast_file_group
        forecast_file_group = pd.concat([forecast_file_group, forecast_file], ignore_index=True)
        # add date to valid dates
        valid_dates.append(date.strftime('%m-%d-%Y'))
    except:
        print(f"File for {date.strftime('%m-%d-%Y')} not found, using previous valid date instead...")
        forecast_file = pd.read_csv(f"//sce.eix.com/workgroup/SPE_DSA/Charles/ADS_forecast/Dist Circuit FPI2 Forecast_{valid_dates[-1]}.csv",)
        # make sure date is in string format
        forecast_file['Date'] = pd.to_datetime(forecast_file['Datetime']).dt.strftime('%Y-%m-%d')
        # keep only current date
        forecast_file = forecast_file[forecast_file['Date'] == date.strftime('%Y-%m-%d')]
        # append to forecast_file_group
        forecast_file_group = pd.concat([forecast_file_group, forecast_file], ignore_index=True)
# sort by Date and FCZ
forecast_file_group = forecast_file_group.sort_values(by=['Circuit','Datetime']).reset_index(drop=True)
```

```python
# create cols to keep for distribution circuits
cols_to_keep_dist = ['Datetime','Date','Circuit', 'FuelsIndex_max_FPI2', 'LFPwtdd_max_FPI2','Max_Wind','Max_Gust']
# concat dist_df with forecast_file_group
dist_df_with_forecast = pd.concat([dist_df[cols_to_keep_dist], forecast_file_group[cols_to_keep_dist]], ignore_index=True)
# save file to Distribution_Forecast_2025.csv
dist_df_with_forecast.to_csv('CircuitLevelData/Dist_Forecasts_2025.csv', index=False)
```
